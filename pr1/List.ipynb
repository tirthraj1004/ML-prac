```
lst = ["x","y","z"]
print(lst)
```

List is ordered

```
n = len(lst)
for i in range(n):
  print("<"+str(i)+">" +lst[i]) 
```

List can have multiple data type

```
lst2 = [1, "ABC", True, 5.5]
print(type(lst2))
for x in lst2:
  print(str(x)+ ": " + str(type(x)))
```

List is mutable

```
lst2[1] = "xyz"
lst2.append(100)
lst2.append(200)
lst2.remove(200)
lst2.insert(3,False)
print(lst2)
```

Slicing list

```
print(lst2[2:5])
```

```
print(lst2[3:])
```

```
print(lst2[:3])
```

Nested list

```
lst3 = [lst,lst2]
print(lst3)
```
