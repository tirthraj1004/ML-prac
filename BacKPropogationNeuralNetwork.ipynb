```python
import numpy as np
```

#Functions

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
  return x*(1-x)

def classify(x):
  if x > 0.5: return 1
  else: return 0
```

#Building Neural Network

```python
input = np.array([[0,0],
                  [0,1],
                  [1,0],
                  [1,1]])

output = np.array([[0],
                   [1],
                   [1],
                   [0]])
```

```python
learningRate = 0.1
inpLayerNeuron = 2
hiddenLayerneuron = 2
outputLayerNeuron = 1
epochs=10000
```

```python
hidden_weights = np.random.uniform(size=(inpLayerNeuron,hiddenLayerneuron))
hidden_bias = np.random.uniform(size=(1,hiddenLayerneuron))

output_weights = np.random.uniform(size=(hiddenLayerneuron,outputLayerNeuron))
output_bias = np.random.uniform(size=(1,outputLayerNeuron))

print("*Hidden Layer* \n\tWeights:",*hidden_weights,"  Bias: ",hidden_bias, end = ' ')
print("\n*Output Layer* \n\tWeights:",*output_weights,"  Bias: ",output_bias, end = ' ')
```

#Training

```python
for x in range(epochs):
  #Forward  
  hidden_layer_activation = np.dot(input, hidden_weights) + hidden_bias
  hidden_layer_out = sigmoid(hidden_layer_activation)

  output_layer_activation = np.dot(hidden_layer_out, output_weights) + output_bias
  Predicted_Output = sigmoid(output_layer_activation)
  
  #Backward
  error = output - Predicted_Output
  d_predicted_output = error * sigmoid_derivative(Predicted_Output)

  error_hidden_layer = d_predicted_output.dot(output_weights.T)
  d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_layer_out)

  #update weight, bias
  output_weights += hidden_layer_out.T.dot(d_predicted_output) * learningRate
  output_bias += np.sum(d_predicted_output, axis = 0, keepdims= True) * learningRate
  hidden_weights += input.T.dot(d_hidden_layer) * learningRate
  hidden_bias += np.sum(d_hidden_layer, axis=0,keepdims=True) * learningRate
```
