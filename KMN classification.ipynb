```
import pandas as pd
import numpy as nm
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap  
import matplotlib.pyplot as mtp
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
```

#Data Set

```
irisData = datasets.load_iris() 
X = irisData.data[:,:2]
Y = irisData.target
```

#Split and Scale

```
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
sclr = StandardScaler()
X_train = sclr.fit_transform(X_train)
X_test = sclr.fit_transform(X_test)
```

#KNN

```
KNNcls = KNeighborsClassifier(n_neighbors=3)
KNNcls.fit(X_train,Y_train)
```

```
def plotData(Neighbour):
  x_set, y_set = X_test, Y_test
  KNNcls.n_neighbors = Neighbour
  Pred_Y = KNNcls.predict(X_test)    
  x1, x2 = nm.meshgrid(nm.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step  =0.01),  
  nm.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))  
  mtp.contourf(x1, x2, KNNcls.predict(nm.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),  
  alpha = 0.65, cmap = ListedColormap(('red','green','blue' )))  
  mtp.xlim(x1.min(), x1.max())  
  mtp.ylim(x2.min(), x2.max())  
  for i, j in enumerate(nm.unique(y_set)):  
      mtp.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], c = ListedColormap(('red', 'green','blue'))(i), label = j, s=50)  
  mtp.title('K-NN Algorithm (with 3 nearest neighbour)')  
  mtp.xlabel('Sepal Length')  
  mtp.ylabel('Sepal Width')  
  mtp.legend()  
```

```
plotData(Neighbour=3)
```

```
plotData(Neighbour=5)
```

```
plotData(Neighbour=7)
```

#Accuracy

```
KNNcls.n_neighbors = 3
Pred_Y = KNNcls.predict(X_test)  
print(confusion_matrix(Y_test,Pred_Y))
print(classification_report(Y_test,Pred_Y))
```

```
KNNcls.n_neighbors = 5
Pred_Y = KNNcls.predict(X_test)  
print(confusion_matrix(Y_test,Pred_Y))
print(classification_report(Y_test,Pred_Y))
```

```
KNNcls.n_neighbors = 7
Pred_Y = KNNcls.predict(X_test)  
print(confusion_matrix(Y_test,Pred_Y))
print(classification_report(Y_test,Pred_Y))
```
